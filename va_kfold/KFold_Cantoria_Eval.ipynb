```python
import sys
import numpy as np
import pandas as pd
import va_plots
import cantoria_utils

np.set_printoptions(threshold=sys.maxsize)
```

```python
mix_df = []
sop_df = []
alto_df = []
ten_df = []
bass_df = []
title = []

for i in range(1, 6):

    mix, s, a, t, b = cantoria_utils.compute_kfold_metrics(train=2, fold=i)
    mix_df.append(mix)
    sop_df.append(s)
    alto_df.append(a)
    ten_df.append(t)
    bass_df.append(b)
    title.append('MaskVoasCNNv2 (K-Fold) - Fold ' + str(i))
```
