```python
import sys
import numpy as np
import pandas as pd
import va_plots
import cantoria_utils

np.set_printoptions(threshold=sys.maxsize)
```

```python
#models_to_compute = [1]
#model_namelist = ["MaskVoasCNNv2"]

models_to_compute = [0, 1, 3, 4, 5]
model_namelist = ["MaskVoasCNN",
                  "MaskVoasCNNv2",
                  "DownsampleVoasCNNv2",
                  "VoasCNN (Retrained)",
                  "VoasCNN (Original)"]

measures = ["Raw Pitch Accuracy",
            "Raw Chroma Accuracy",
            "Overall Accuracy",
            "Voicing Recall",
            "Voicing False Alarm"]

agg_metrics = pd.DataFrame()

for model_index, model_name in zip(models_to_compute, model_namelist):
    agg_model = pd.DataFrame()
    mix_df, sop_df, alto_df, ten_df, bass_df = cantoria_utils.compute_holdout_metrics(model_index)
    sop_agg = sop_df[measures].agg(['mean', 'std', 'median'])
    alto_agg = alto_df[measures].agg(['mean', 'std', 'median'])
    ten_agg = ten_df[measures].agg(['mean', 'std', 'median'])
    bass_agg = bass_df[measures].agg(['mean', 'std', 'median'])
    sop_agg["Voice"] = np.repeat(["Soprano"], sop_agg.shape[0])
    alto_agg["Voice"] = np.repeat(["Alto"], alto_agg.shape[0])
    ten_agg["Voice"] = np.repeat(["Tenor"], ten_agg.shape[0])
    bass_agg["Voice"] = np.repeat(["Bass"], bass_agg.shape[0])
    agg_model = pd.concat([sop_agg, alto_agg, ten_agg, bass_agg])
    agg_model["Model"] = np.repeat([model_name], agg_model.shape[0])
    agg_metrics = pd.concat([agg_metrics, agg_model])

agg_metrics.reset_index(inplace=True)
agg_metrics.rename(columns={"index": "Dist Measure"}, inplace=True)
```

```python
agg_proc = agg_metrics.melt(id_vars=["Model", "Voice", "Dist Measure"], var_name="MIR Measure")
#agg_proc = agg_proc[agg_proc["Dist Measure"].isin(['mean', 'std'])]
agg_proc = agg_proc.pivot(index=["MIR Measure", "Model"], columns=["Voice", "Dist Measure"])
agg_proc = agg_proc['value'].reset_index()
agg_proc = agg_proc.set_index(["MIR Measure", "Model"])
agg_proc
```
